```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm 
mpl.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('health_data.csv')

path = '/content/NanumGothicEco.ttf'  
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)

plt.rc('font', family=font_name)
```

```python
#smoking status - bar chart

fm._rebuild()

plt.title('\nSmoking status')
df['흡연상태'].value_counts()
df['흡연상태'].value_counts().plot.bar(rot = 0)
```

#smoking status - bar chart
현재 비흡연자(No+NotNow)의 비율이 압도적으로 높게 보인다. 이를 통해 흡연 문화는 현재 분명한 비주류 문화임을 알 수 있다.



```python
#drinking status - pie chart

fm._rebuild()

plt.title("\nMen's drinking status")
malc = df[df['성별코드']=='M'].iloc[:,26]
malc.value_counts().plot(kind='pie')
plt.show()

plt.title("\nWomen's drinking status")
falc = df[df['성별코드']=='F'].iloc[:,26]
falc.value_counts().plot(kind = 'pie')
plt.show()
```

#drinking status - pie chart
남성의 음주 비율이 여성보다 월등히 높음을 알 수 있다. 이는 남성의 음주로 인한 질병 발생에 각별히 신경 써야 함을 시사한다.

```python
#cholesterol - scatter chart

fm._rebuild()

total = df.iloc[:,15]
hdl = df.iloc[:,17]
ldl = df.iloc[:,18]

total_hdl_df = pd.DataFrame({'total': total, 'hdl': hdl})
plt.scatter(total_hdl_df['total'], total_hdl_df['hdl'])
plt.title("\nTotal Cholesterol-HDLCholesterol")
plt.xlabel('Total Cholesterol')
plt.ylabel('HDLCholesterol')
plt.show()    

total_ldl_df = pd.DataFrame({'total': total, 'ldl': ldl})
plt.scatter(total_ldl_df['total'], total_ldl_df['ldl'])
plt.title("\nTotal Cholesterol-LDLCholesterol")
plt.xlabel('Total Cholesterol')
plt.ylabel('LDLCholesterol')
plt.show()
```

#cholesterol - scatter chart
HDL콜테스테롤은 나쁜 콜레스테롤을 없애는 성분이고 LDL콜레스테롤은 각종 질병을 야기하는 성분이다. 총콜레스테롤과 HDL콜레스테롤은 서로 별다른 관계가 보이지 않으나, 총콜레스테롤과 LDL콜레스로롤은 서로 비례하는 관계를 보인다. 이를 통해 콜레스테롤 수치 증가가 가진 위험성은 LDL콜레스테롤에서 비롯됨을 유추할 수 있다.
